### Basic Product Information

```python
page = requests.get(link)
```

```python
print(page)
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.name
```

```python
names = soup.find_all('span',class_='a-profile-name')
names
```

```python
len(names)
```

```python
cust_name =[]
for i in range (0,len(names)):
    cust_name.append(names[i].get_text())
```

```python
cust_name
```

```python
cust_name.pop(0)
```

```python
cust_name.pop(9) ##removing duplicate name with pop method
```

#### Find all

```python
names = soup.find_all('span',class_='a-size-base review-title a-text-bold')
names
```

#### Review Title

```python
title=soup.find_all(class_='review-title-content')
```

```python
title
```

```python
review_title =[]
for i in range (0,len(title)):
    review_title.append(title[i].get_text())
```

```python
review_title
```

```python
review_title[:] =[titles.lstrip('5.0 out of 5 stars\n')for titles in review_title]
```

```python
review_title[:] =[titles.rstrip('\n,')for titles in review_title]
```

```python
review_title
```

```python
review_title[:] =[titles.strip('\n')for titles in review_title]
```

```python
review_title
```
